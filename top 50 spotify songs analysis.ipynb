This is the very interesting part,where we have to choose a model that we woild love to use for a good prediction

CHOOSING A MODEL.....


```python

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectKBest, f_classif



le = LabelEncoder()

for col in df.columns.values:
  if df[col].dtypes == 'object':
    le.fit(df[col].values)
    df[col] = le.transform(df[col])

df.head()
```

****Here we are selecting the best features,and having  good spice of it is doing that in an ascending order****

```python
X = df.drop('Loudness', axis=1)
y = df.Loudness

selector = SelectKBest(score_func=f_classif, k=5)
fitted = selector.fit(X, y)
features_scores = pd.DataFrame(fitted.scores_)
features_columns = pd.DataFrame(X.columns)

best_features = pd.concat([features_columns, features_scores], axis=1)
best_features.columns = ['Feature', 'Score']
best_features.sort_values(by='Score', ascending=False, inplace=True)
best_features
```

Time to split the data,the data is needed to be splitted to know our test and train

```python
X.drop('Artist_Name', axis=1, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
X_train
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

type(X_train), type(X_test)
```

 i chose to apply K-Means Clustering.to see what i will get as my mean squared error 

```python
clustering = KMeans(n_clusters=2)
clustering.fit(X_train, y_train)

y_pred = clustering.predict(X_test)
print(mean_squared_error(y_test, y_pred))
```

#the mean squared error is not good enough,so defintely we would need to go with another model.

LINEAR REGRESSION MODEL**

```python
le_reg=LinearRegression()
le_reg.fit(X_train,y_train)
predict_linear_reg=le_reg.predict(X_test)
```

```python
from math import sqrt
sqrt(mean_squared_error(predict_linear_reg,y_test))
```

```python
result=pd.DataFrame({'actual':y_test,'cal':predict_linear_reg,'diff':abs(y_test-predict_linear_reg)})
```

```python
result.head(5)
```

#the root mean error is perfect enough,but we can still try to see if other models are prefect enough like this

This is when we decide to apply or try support vector machine (SVM)

```python
regressor = SVR(C=0.5)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
print(mean_squared_error(y_test, y_pred))
```

i think this mean squared error of this particular model is actually fine
